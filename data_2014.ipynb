```python
dateparse = lambda x: pd.datetime.strptime(x, '%d/%m/%Y')

data14 = pd.read_csv("cpcb_dly_aq_delhi-2012.csv", parse_dates=['Sampling Date'], date_parser=dateparse).drop("Location of Monitoring Station", axis=1).sort_values("Sampling Date")    
```

```python
data14["DateMonth"]=data14["Sampling Date"].dt.to_period("M")
```

```python
data14 = data14.groupby("DateMonth").mean()
```
